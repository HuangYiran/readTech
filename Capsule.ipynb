## 简介
capsule是用于取代neuron的一种尝试。与neuro输出一个value不同，每个capsule会输出一个vector。另外和neuron的，只针对一个pattern不同，capsule针对的是一个种类的pattern。其输出vector中每个dimension(这个vector的方向)会代表对象的一种特征，而这个vector的norm，则代表这个种类的pattern是否存在。<br>
